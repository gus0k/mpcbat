```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from structure import build_price
```

```python
def plot_cost_function(slope, breakpoints):
    offset = build_price(slope, breakpoints)
    print('The offsets are:', offset)
    m = breakpoints.min()
    m = m * 2 if m != 0 else -3
    M = breakpoints.max()
    M = M * 2 if M != 0 else 3
    X = np.arange(m, M, 0.1).reshape(1, -1)
    f = np.dot(slope.reshape(-1, 1), X) + offset.reshape(-1, 1)
    fig, ax = plt.subplots()
    ax.plot(X.reshape(-1, 1), f.max(axis=0))
    ax.axhline(0, c='k')
    ax.axvline(0, c='k')
    for b in breakpoints:
        ax.axvline(b, c='r')
```

```python
slope = np.array([1, 3, 6, 10])
breakpoints = np.array([-3, 4, 5])
plot_cost_function(slope, breakpoints)
```

```python
slope = np.array([1, 3, 6, 10])
breakpoints = np.array([-3, 0, 5])
plot_cost_function(slope, breakpoints)
```

```python
slope = np.array([1, 3])
breakpoints = np.array([0])
plot_cost_function(slope, breakpoints)
```

```python
slope = np.array([1, 2, 3, 4])
breakpoints = np.array([-0.5, 0, 0.7])
plot_cost_function(slope, breakpoints)
```

```python
slope = np.array([1, 2, 4, 4])
breakpoints = np.array([-0.5, 0, 0.7])
plot_cost_function(slope, breakpoints)
```

```python
slope = np.array([1, 2, 4, 4])
breakpoints = np.array([-0.5, 0, 0])
plot_cost_function(slope, breakpoints)
```
